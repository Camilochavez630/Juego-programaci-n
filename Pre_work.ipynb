```python
# Ejercicio 3
# prompt: Crea un histograma a partir de un array aleatorio que siga una distribucin normal N(2,1.5)

import matplotlib.pyplot as plt
import numpy as np

# Genera datos aleatorios con distribucin normal N(2, 1.5)
data = np.random.normal(loc=2, scale=1.5, size=1000)

# Crea el histograma
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histograma de datos con distribucin N(2, 1.5)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Ejercicio 4
# prompt: Crea un DataFrame a partir del dataset de Titanic y muestra las distribuciones de la edad y del importe de los billetes

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset del Titanic
titanic = sns.load_dataset('titanic')

# Crear los histogramas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(titanic['age'].dropna(), kde=True)
plt.title('Distribucin de la Edad')

plt.subplot(1, 2, 2)
sns.histplot(titanic['fare'].dropna(), kde=True)
plt.title('Distribucin del Importe del Billete')

plt.tight_layout()
plt.show()
```
